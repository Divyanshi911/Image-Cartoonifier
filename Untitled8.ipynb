CREATE EDGE MASK

```python
def edge_mask(img, line_size,blur_value):
  gray= cv.cvtColor(img,cv.COLOR_RGB2GRAY)
  gray_blur= cv.medianBlur(gray, blur_value)

  edges=cv.adaptiveThreshold(gray_blur,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,line_size,blur_value)

  return edges
```

```python
line_size,blur_value=7,7
edges= edge_mask(cv.imread(filename), line_size,blur_value)

plt.imshow(edges,cmap='binary')
plt.show()
```

REDUCE THE COLOR PALETTE

```python
def color_quantization(img,k):
  data=np.float32(img).reshape((-1,3))

  criteria=(cv.TERM_CRITERIA_EPS+ cv.TERM_CRITERIA_MAX_ITER, 20,0.001)

  #IMPLEMENTING K-MEANS CLUSTERING#
  ret,label,centre= cv.kmeans(data,k , None, criteria, 10 ,cv.KMEANS_RANDOM_CENTERS )
  centre=np.uint8(centre)

  result= centre[label.flatten()]
  result= result.reshape(img.shape)

  return result    
```

```python
img_quantiz=color_quantization(img,k=6)
plt.imshow(img_quantiz)
plt.show()
```

```python
# reduce the noise

blurred= cv.bilateralFilter(img_quantiz,d=5, sigmaColor=200, sigmaSpace=200)
plt.imshow(blurred)
plt.show()
```

COMBINE EDGE MASK WITH QUANTIZ IMAGE

```python
def cartoon():
  c=cv.bitwise_and(blurred, blurred, mask=edges)

  plt.imshow(c)
  plt.title("Cartoonified Image")
  plt.show()

  plt.imshow(org_img)
  plt.title("Original Image")
  plt.show()
```
